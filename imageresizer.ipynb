```python
# ==============================
# IMAGE RESIZER TOOL (Task 7)
# ==============================

# Install Pillow (if not installed)
!pip install pillow

import os
from PIL import Image

# -------- USER SETTINGS --------
input_folder = "/content/sample_data/input_images"      # Folder containing original images
output_folder = "/content/sample_data/output_images"    # Folder to save resized images
new_width = 800
new_height = 800
output_format = "PNG"   # Change to "JPEG" or "PNG"
# --------------------------------

# Create output folder if not exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Supported formats
valid_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

# Process Images
count = 0

for filename in os.listdir(input_folder):
    if filename.lower().endswith(valid_extensions):
        try:
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path)

            # Resize image
            resized_img = img.resize((new_width, new_height))

            # Change file name & format
            new_name = os.path.splitext(filename)[0] + "." + output_format.lower()
            save_path = os.path.join(output_folder, new_name)

            resized_img.save(save_path, output_format)

            count += 1
            print(f"Processed: {filename}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("\n✅ Done!")
print(f"Total images processed: {count}")
print(f"Resized images saved in: {output_folder}")
```
